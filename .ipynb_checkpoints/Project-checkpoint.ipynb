We plan to combine a subset of each of thes two data sets in order to explore relationships between team performance and attendance.

Attendance Data: ESPN has public data of a Team's Home/Away/Overall attendance from 2000-Present day. This isn't already formatted as we will have to collect it all and clean it to our needs for analyzing. [One year of data]('http://www.espn.com/nhl/attendance/_/year/2019')

We'll have to join these two datasets together in some way, this will be done via year and team name. Knowing the 'Teams.csv' data only goes up to 2011, and the ESPN data starts 2000 we will have completed data from the years 2000-2011. 

#### Objectives for analyzing your dataset

Our goal is to find a relationship between and, ideally, predict attendance based on one or more of the following factors:

## Hypotheses/Problem
**Null**: An NHL teams' point totals have no effect on average home attendance  
**Alternative**: NHL teams' point totals have an effect on average home attendance
  
Along with this hypothesis, we want to be able to predict a team's average attendance based on their stats and performance. In order to accomplish this, we will create a linear regression model.

```python
# First Ten Records of Datset

import pandas as pd
#import matplotlib.pyplot as plt
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

df = pd.read_csv('Teams.csv')

df = df.drop(df[df.year < 2000].index)
df = df.drop(df[df.lgID != "NHL"].index)
df = df.drop('lgID', 1)

for index, row in df.iterrows():
    if row['playoff'] not in ['CQF', 'CSF', 'CF', 'F', 'SC']:
        df.at[index, 'playoff'] = 'None'

df['year'] = df['year'].astype('category')
df['tmID'] = df['tmID'].astype('category')
df['confID'] = df['confID'].astype('category')
df['divID'] = df['divID'].astype('category')
df['playoff'] = df['playoff'].astype('category')
df['name'] = df['name'].astype('category')
df['rank'] = df['rank'].astype('category')

df = df.drop('franchID', 1)

attendance_df = pd.read_csv('attendance.csv')
# attendance_df.info()

def pull_city(full_name) :
    if ('New York' in full_name):
        name_split = full_name.split()
        return "NY " + name_split[2]
    elif ('New ' in full_name or 'St. ' in full_name or 'Los ' in full_name or 'San ' in full_name or 'Tampa ' in full_name):
        name_split = full_name.split()
        return name_split[0] + ' ' + name_split[1]
    elif (' of ' not in full_name) :
        return full_name.split()[0]
    else :
        name_split = full_name.split()
        return name_split[len(name_split) - 1]



df['name'] = df['name'].map(lambda name: pull_city(name))
```

```python
attendance_df = attendance_df.drop('Home Percentage', 1)
attendance_df = attendance_df.drop('Road Percentage', 1)
attendance_df = attendance_df.drop('Percentage', 1)

# Replace Arizona with Phoenix and replace Winnipeg with Atlanta
for index, row in attendance_df.iterrows():
    if row['Team'] == 'Arizona':
        attendance_df.at[index, 'Team'] = 'Phoenix'
    elif row['Team'] == 'Winnipeg' and row['Year'] != 2011:
        attendance_df.at[index, 'Team'] = 'Atlanta'

       

attendance_df['Rank'] = attendance_df['Rank'].astype('category')
attendance_df['Team'] = attendance_df['Team'].astype('category')
attendance_df['Year'] = attendance_df['Year'].astype('category')
attendance_df = attendance_df.rename(columns={"Rank": 'Attendance Rank'}) 

combined_df = pd.merge(df, attendance_df,  how='right', left_on=['name','year'], right_on = ['Team','Year'])
combined_df = combined_df.drop('Year', 1)
combined_df = combined_df.drop('Games', 1)
combined_df = combined_df.drop('Team', 1)

combined_df = combined_df.rename(columns={"rank": "division rank"})
```

# Data Preprocessing
Luckily for us, the NHL data collection is pretty thorough and doesn't leave many missing values. However, there is one exception due to a rule change before the 2005 season. Previously, the NHL did not have any type of overtime and games could end in a tie. Starting in 2005, a 5 minute sudden-death overtime would be played in the event of a tie. If neither team scored in the 5 minutes, a shootout would then decide the game. This results in no values for overtime losses or shootout wins/losses prior to 2005. Additionally, only years prior to 2005 have values for ties.  
Fortunately, ties and and overtime losses are both worth 1 point. This means we can use points as an independent variable that is more or less standardized when it comes to this discrepency.  
  
While we didn't have to worry about many missing values, we had to do some cleaning to make sure that we could properly combine our 2 data sets. We joined the data by team name and year into a single dataframe. However, our team stats data listed team names as the full name (ex. Philadelphia Flyers) while the attendance data listed teams just by the city (Philadelphia). We ended up writing a function that pulls the city names out of the full name and mapped the original team stats dataframe to be consistent with our attendance data. We also had to deal with a couple cases where teams moved and/or changed their name. In 2011, the Atlanta Thrashers moved to Winnipeg and changed their name to the Winnipeg Jets. This was an issue because the attendance data listed rows for the Atlanta Thrasers under their new name (Winnipeg) even for years prior to the name change. Similarly, the Phoenix Coyotes changed their name to be the Arizona Coyotes in 2014. The attendance data once again listed all entries under the new name (Arizona) while the team data listed each of their entries under the name that they actually had at the time (Phoenix). Lucky for us, these were pretty simple fixes.

### Correlation  Coefficients

```python
from scipy.stats import pearsonr
def print_corr_and_p(display_name, col_name):
    corr_and_r = pearsonr(combined_df[col_name], combined_df['Home Average Attendance'])
    print(display_name)
    print('\t' + 'Correlation Coeff.: ' + str(corr_and_r[0]))
    print('\t' + 'p-value: ' + str(corr_and_r[1]))
    print()
    
print_corr_and_p("Points", "Pts")
print_corr_and_p("Goals For", "GF")
print_corr_and_p("Penalty Minutes", "PIM")
```

For goals for and points we see moderate positive correlations. Relating this back to out hypothesis, our p-value is far less than a 0.03 significant level. This means that we reject the null hypothesis as we have evidence that points do have some effect on a team's average home attendance.

# Data Modeling and Prediction
As stated before, we will be using points, goals for, and penalty minutes to try and predict attendance. This will be our first model since it includes the 3 main variables we picked from the beginning. We will then move on to some exploratory models.



```python
import numpy as np
from sklearn.linear_model import SGDRegressor
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import Imputer
from sklearn.preprocessing import StandardScaler

# function from tutorial
def create_feature_matrices(df_train, df_test, selected_columns, dummy_columns=None):
    """
    Creates feature matrices for the training and testing sets from the given dataframes.
    The feature matrices are built from the columns given in selected columns.
    """
    
    if dummy_columns is None:
        dummy_columns = dict()
    
    # keep track of numerical features
    numerical_trial_columns = []
    
    # keep track of dummy features for categorical variables
    categorical_trial_columns = []
    
    # build feature lists
    for col_name in selected_columns:
        if col_name in dummy_columns:
            categorical_trial_columns.extend(dummy_columns[col_name])
        else:
            numerical_trial_columns.append(col_name)

    # transform numerical features
    if len(numerical_trial_columns) > 0:
        X_train_numerical = df_train[numerical_trial_columns].astype("float64").values
        X_test_numerical = df_test[numerical_trial_columns].astype("float64").values
    
        imputer = Imputer(missing_values=0)
        X_train_numerical = imputer.fit_transform(X_train_numerical)
        X_test_numerical = imputer.transform(X_test_numerical)
    
        scaler = StandardScaler()
        X_train_numerical = scaler.fit_transform(X_train_numerical)
        X_test_numerical = scaler.transform(X_test_numerical)
    
    # select categorical features
    if len(categorical_trial_columns) > 0:
        X_train_categorical = df_train[categorical_trial_columns].astype("float64").values
        X_test_categorical = df_test[categorical_trial_columns].astype("float64").values
    
    # concatenate feature matrices
    if len(numerical_trial_columns) > 0 and len(categorical_trial_columns) > 0:
        X_train = np.hstack([X_train_numerical, X_train_categorical])
        X_test = np.hstack([X_test_numerical, X_test_categorical])
    elif len(numerical_trial_columns) > 0:
        X_train = X_train_numerical
        X_test = X_test_numerical
    else:
        X_train = X_train_categorical
        X_test = X_test_categorical
        
    return X_train, X_test

def plot_predictions(predicted):
    plt.figure(figsize=(10, 8))
    sns.regplot(x=y_test, y=predicted, fit_reg=False)
    plt.plot([0, np.max(y_test)], [0, np.max(predicted)])
    plt.xlabel("True Attendance", fontsize=16)
    plt.ylabel("Predicted Attendance", fontsize=16)
```

```python
columns = ['Pts', 'GF', 'PIM']

df_train, df_test = train_test_split(combined_df)

X_train, X_test = create_feature_matrices(df_train, df_test, columns)

y_train = df_train["Home Average Attendance"].values
y_test = df_test["Home Average Attendance"].values

lr = SGDRegressor(max_iter=5000)
lr.fit(X_train, y_train)
predicted_test = lr.predict(X_test)

mse = mean_squared_error(y_test, predicted_test)
rmse = np.sqrt(mse)
print()
print("Model MSE:", mse)
print("Model RMSE:", rmse)
```

```python
plot_predictions(predicted_test)
```

To contrast the previous model, we will be adding in more of the variables. We don't really think that these vairables are very predictive of attendance but we're going to explore the possibility regardless.

```python
numerical_columns = ["Pts", "GF", "PIM", "PPG", "PPC", "SHA", "PKG", "SHF"]
categorical_columns = ["divID", "division rank", "playoff", "tmID"]
selected_columns = numerical_columns + categorical_columns

dummy_columns = {}
for col_name in categorical_columns:
    dummies = pd.get_dummies(combined_df[col_name], prefix=col_name)
    dummy_columns[col_name] = dummies.columns.values
    combined_df = combined_df.merge(dummies, left_index=True, right_index=True)
    combined_df = combined_df.drop(columns=col_name)

df_train, df_test = train_test_split(combined_df)

X_train, X_test = create_feature_matrices(df_train, df_test, numerical_columns, dummy_columns=dummy_columns)

y_train = df_train["Home Average Attendance"].values
y_test = df_test["Home Average Attendance"].values

lr = SGDRegressor(max_iter=5000)
lr.fit(X_train, y_train)
predicted_with_cat = lr.predict(X_test)

mse = mean_squared_error(y_test, predicted_with_cat)
rmse = np.sqrt(mse)
print()
print("Model MSE:", mse)
print("Model RMSE:", rmse)
```

```python
plot_predictions(predicted_with_cat)
```

While this model wasn't far off from the first, there was more error. For our last model, we will create a model for each of our variables and use their sorted mean square error values to sort them. Starting with the model with the least amount of error, we will add another variable to the model, if the error goes down we will keep it, if not, we will throw it out and move on.

## Greedy Approach

```python
def sort_columns(df_train, df_test, y_train, y_test, all_columns, dummy_columns=None):
    column_rmses = []

    for col_name in all_columns:
        X_train, X_test = create_feature_matrices(df_train,
                                                 df_test,
                                                 [col_name],
                                                 dummy_columns=dummy_columns)
    
        simple_lr = SGDRegressor(max_iter=5000)
        simple_lr.fit(X_train, y_train)
        pred_prices_simple_test = simple_lr.predict(X_test)
        simple_mse = mean_squared_error(y_test, pred_prices_simple_test)
        simple_rmse = np.sqrt(simple_mse)
        print()
        print(col_name + " Model MSE:", simple_mse)
        print(col_name + " Model RMSE:", simple_rmse)
        column_rmses.append((simple_rmse, col_name))
    
    column_rmses.sort(key=lambda p: p[0])
    sorted_columns = [col_name for _, col_name in column_rmses]
    print()
    print("Sorted columns:" + str(sorted_columns))
    
    return sorted_columns

def select_features(df_train, df_test, y_train, y_test, sorted_columns, dummy_columns=None):
    lr = SGDRegressor(max_iter=5000)
    
    train_zeros = np.zeros((y_train.shape[0], 1))
    test_zeros = np.zeros((y_test.shape[0], 1))
    lr.fit(train_zeros, y_train)
    pred_y_test = lr.predict(test_zeros)
    
    null_mse = mean_squared_error(y_test, pred_y_test)
    null_rmse = np.sqrt(null_mse)
    
    base_model_columns = []
    base_rmse = null_rmse
    base_mse = null_mse

    print("Base Model MSE:", base_mse)
    print("Base Model RMSE:", base_rmse)
    print()

    for col_name in sorted_columns:
        # track all column names
        trial_columns = base_model_columns[:]
        trial_columns.append(col_name)

        X_train, X_test = create_feature_matrices(df_train,
                                                  df_test,
                                                  trial_columns,
                                                  dummy_columns=dummy_columns)
    
        lr = SGDRegressor(max_iter=5000)
        lr.fit(X_train, y_train)
        pred_y_test = lr.predict(X_test)
        trial_mse = mean_squared_error(y_test, pred_y_test)
        trial_rmse = np.sqrt(trial_mse)
        print()
        print(str(trial_columns) + " Model MSE:", trial_mse)
        print(str(trial_columns) + " Model RMSE:", trial_rmse)
    
        if trial_rmse < base_rmse:
            print("Updating base model")
            base_rmse = trial_rmse
            base_mse = trial_mse
            base_model_columns = trial_columns
        
        print()
        print("Base Model MSE:", base_mse)
        print("Base Model RMSE:", base_rmse)
        print()
    
    print("Our chosen columns are: " + str(base_model_columns))
    return base_model_columns
```

```python
sorted_columns = sort_columns(df_train, df_test, y_train, y_test, selected_columns, dummy_columns)
```

An interesting variable to include is tmID (team name). This essentially means the model will be considering the team's average attendance of previous years. It proved to be the best predictor of attendance on its own. We would guess this is because there are a lot of other factors that contribute to attendance outside of hockey entirely. Certain cities simply have more people and/or embrace their hockey teams much more than others. Maybe other major professional sports such as an NBA or NFL team in the same city pull some attention away from the NHL. Including the team in the model indirectly accounts for these factors. However, it does limit us in some way. If we have team id in tthis model, we won't be able to punch in some theoretical stats without an designating a team.

```python
selected_columns = select_features(df_train, df_test, y_train, y_test, sorted_columns, dummy_columns=dummy_columns)

X_train, X_test = create_feature_matrices(df_train, df_test, selected_columns, dummy_columns)


lr = SGDRegressor(max_iter=5000)
lr.fit(X_train, y_train)
greedy_test = lr.predict(X_test)

mse = mean_squared_error(y_test, greedy_test)
rmse = np.sqrt(mse)
```

```python
plot_predictions(greedy_test)
```

With the greedy approach we produced the best model yet. Additionally, two of our predicted features (points and goals for) made it into the greedy model.  
Overall, we produced a model with an RSME of 1263.91. Let's take a quick look at the distribution of average home attendance to get a better idea of the significance of this number.

Relative to the average attendance counts, being off by 1263 isn't necessarily small but it isn't very large either.
